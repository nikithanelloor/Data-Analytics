```python
## Votes Difference between Restaurants that Deliver and Do not Deliver

# Filter data for restaurants that provide online delivery and those that do not
yes_delivery_votes = df_cleaned[df_cleaned['Has Online delivery']=='Yes']['Votes']
no_delivery_votes = df_cleaned[df_cleaned['Has Online delivery']=='No']['Votes']

## calculate avg number of votes
avg_yes_delivery_votes=yes_delivery_votes.mean()
avg_no_delivery_votes=no_delivery_votes.mean()

## Calculate the difference
votes_diff=avg_yes_delivery_votes-avg_no_delivery_votes

print(f"yes delivery votes : {yes_delivery_votes}")
print(f"no delivery votes : {no_delivery_votes}")

print(f"avg yes delivery votes : {avg_yes_delivery_votes:.2f}")
print(f"avg no delivery votes : {avg_no_delivery_votes:.2f}")

print(f"Difference in average votes: {votes_diff:.2f}")
```

```python
## Top 10 Cuisines Served Across Cities
all_cuisines=df_cleaned['Cuisines'].str.split(',').explode()  # split cuisines by comma and create a list of all cuisines
all_cuisines
```

```python

df_cleaned['Cuisines Count']=df_cleaned['Cuisines'].apply(lambda x:len(x.split(',')) if isinstance(x,str)else 0)
print(df_cleaned[['Restaurant ID','Cuisines','Cuisines Count']].head())
```

```python
# clean up the cuisine names # removeleading/trailing spaces and esure single comma and space between cuisines
def clean_cuisine(cuisine):
    cleaned_cuisine=', '.join([c.strip() for c in cuisine.split(',')])  
    return cleaned_cuisine
    
# Apply the cleaning function to 'Cuisines' column
df_cleaned['Cuisines']= df_cleaned['Cuisines'].apply(lambda x: clean_cuisine(x))

# Check cleaned 'Cuisine' column
print(df_cleaned[['Restaurant ID','Cuisines']].head(10))

# Flatten 'Cuisine' column into list of all cuisines
all_cuisines=[Cuisine for Cuisines in df_cleaned['Cuisines']  for Cuisine in Cuisines.split(',')]  # split cuisines by comma and create a list of all cuisines

top_cuisines=pd.Series(all_cuisines).value_counts().head(10)
print(top_cuisines)
```

```python
## Top 10 cuisines
# Split the cuisines into separate rows if there are multiple cuisines in one entry
df_expanded = df_cleaned['Cuisines'].str.split(',').explode().str.strip()

# Count the occurrences of each cuisine across all cities
cuisine_count = df_expanded.value_counts().reset_index()

# Rename columns for clarity
cuisine_count.columns = ['Cuisine', 'Cuisine Count']

# Get the top 10 cuisines
top_10_cuisines = cuisine_count.head(10)

# Display the result
print(top_10_cuisines)
```

```python
#  What is the maximum and minimum number of cuisines that a restaurant serves? Also, which is the most served cuisine across the restaurant for each city?

df_no_of_cuisines=df_cleaned['Cuisines'].str.split(',').apply(len)
df_no_of_cuisines
max_cuisinies=df_no_of_cuisines.max()
min_cuisinies=df_no_of_cuisines.min()
print(f"Maximum no.of cuisines a restaurant serves: ",max_cuisinies)
print(f"Miniimum no.of cuisines a restaurant serves: ",min_cuisinies)
```

```python
df_expanded2=df_cleaned[['City','Cuisines']]
# Reset the index to ensure unique indices and drop the old index
df_expanded2 = df_expanded2.reset_index(drop=True)

# Split the 'Cuisines' column by commas, explode into multiple rows, and strip whitespace
df_expanded2['Cuisines']=df_expanded2['Cuisines'].str.split(',')

# Melt the DataFrame to convert the list of cuisines into separate rows
df_expanded2 = df_expanded2.explode('Cuisines')

# Strip any leading/trailing whitespace from the cuisines
df_expanded2['Cuisines'] = df_expanded2['Cuisines'].str.strip()

df_expanded2
```

```python
# Group by city and cuisine to count the occurrences
city_cuisine_count = df_expanded2.groupby(['City', 'Cuisines']).size().reset_index(name='Cuisine Count')
city_cuisine_count
```

```python
# For each city, find the most served cuisine
most_served_cuisine_per_city = city_cuisine_count.loc[city_cuisine_count.groupby('City')['Cuisine Count'].idxmax()]
print("Most served cuisines across cities:")
most_served_cuisine_per_city.head(10)
```

```python
 # Factors That May Affect Ratings

# Checking correlation between factors like cuisines, cost, delivery options, votes, and ratings
rating_factors = df_cleaned[['Average Cost for two', 'Cuisine Count', 'Has Online delivery', 'Has Table booking', 'Votes', 'Price range', 'Aggregate rating']]
# Convert categorical data to numeric for correlation analysis
# Use .loc to explicitly modify the DataFrame in place
rating_factors.loc[:, 'Has Online delivery'] = rating_factors['Has Online delivery'].map({'Yes': 1, 'No': 0})
rating_factors.loc[:, 'Has Table booking'] = rating_factors['Has Table booking'].map({'Yes': 1, 'No': 0})

# Check the updated DataFrame
print(rating_factors.head())

```

```python
# Calculate correlations
correlation_matrix = rating_factors.corr()

# Display correlation matrix
print("\nCorrelation matrix showing factors affecting ratings:")
print(correlation_matrix)

# Heatmap of correlations
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Factors Affecting Ratings")
plt.show()
```
